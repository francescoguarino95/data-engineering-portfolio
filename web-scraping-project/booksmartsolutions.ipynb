# Scraping delle pagine di una libreria online

Questo progetto ha l’obiettivo di estrarre dati dal sito books.toscrape.com
tramite tecniche di web scraping. Le informazioni raccolte per ogni libro
sono titolo, rating, prezzo e disponibilità, e vengono poi salvate
in un file CSV per successive analisi.

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
books = []

# Ciclo sulle 50 pagine di catalogo del sito
for page in range(1, 51):
    url = f"https://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    items = soup.find_all("article", class_="product_pod")

    for item in items:
        # Titolo del libro
        title = item.h3.a["title"]

        # Valutazione (converto le parole in numeri)
        rating = item.p["class"][1]
        rating_map = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}
        stars = rating_map.get(rating, 0)

        # Prezzo
        price = item.find("p", class_="price_color").text.strip()

        # Disponibilità
        availability = item.find("p", class_="instock availability").text.strip()

        # Aggiungo il libro alla lista
        books.append([title, stars, price, availability])
```

```python
# Salvo tutto in un file CSV
with open("books.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Titolo", "Valutazione (stelle)", "Prezzo", "Disponibilità"])
    writer.writerows(books)

print("Scraping completato, file 'books.csv' creato con successo.")
```
