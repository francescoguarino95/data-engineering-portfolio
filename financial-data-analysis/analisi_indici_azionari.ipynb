# Analisi dell'Andamento degli Indici Azionari S&P 500 ed EURO STOXX 50

Questo progetto analizza l’andamento di due indici azionari globali:
S&P 500 e EURO STOXX 50. Utilizzando Python e Pandas, vengono analizzati
i dati storici degli ultimi anni per calcolare i rendimenti percentuali,
individuare i giorni di maggiore volatilità e studiare il volume medio
degli scambi.
L’obiettivo è ottenere alcune informazioni utili sull’andamento dei mercati
e sui possibili pattern legati ai giorni della settimana o ai periodi dell’anno.

```python
import pandas as pd

# CARICHIAMO I DATI
def load_data(path):
    df = pd.read_csv(path)  # leggo il file CSV
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce", utc=True)  # converto la colonna "Date" in formato data
    df = df.set_index("Date").sort_index()  # imposto "Date" come indice e ordino cronologicamente
    return df

# FUNZIONE PER EFFETTUARE L'ANALISI
def analyze_index(df, name):
    print(f"\nAnalisi {name}")

    # RENDIMENTO PERCENTUALE MENSILE E ANNUALE
    monthly_return = df["Close"].resample("ME").last().pct_change() * 100  # prendo l’ultimo valore del mese e calcolo la variazione %
    annual_return = df["Close"].resample("YE").last().pct_change() * 100  # stessa cosa ma a livello annuale
    print("\nRendimento mensile (ultimi 5 valori):")
    print(monthly_return.tail())  # stampo solo gli ultimi 5 del rendimento mensile
    print("\nRendimento annuale (ultimi 5 valori):")
    print(annual_return.tail())  # stampo solo gli ultimi 5 del rendimento annuale

    # RENDIMENTO MEDIO GIORNALIERO PER GIORNO DELLA SETTIMANA
    df["DailyReturn"] = df["Close"].pct_change() * 100  # creo una nuova colonna con la variazione giornaliera %
    mean_by_day = df.groupby(df.index.day_name())["DailyReturn"].mean()  # calcolo la media dei rendimenti per ogni giorno della settimana
    print("\nRendimento medio per giorno della settimana:")
    print(mean_by_day)

    # GIORNO CON RENDIMENTO MASSIMO E MINIMO
    max_day = df["DailyReturn"].idxmax()  # trovo la data con rendimento più alto
    min_day = df["DailyReturn"].idxmin()  # trovo la data con rendimento più basso
    print(f"\nGiorno con rendimento massimo: {max_day.date()} ({df.loc[max_day,'DailyReturn']:.2f}%)")
    print(f"Giorno con rendimento minimo: {min_day.date()} ({df.loc[min_day,'DailyReturn']:.2f}%)")

    # VOLUME MEDIO GIORNALIERO
    print(f"\nVolume medio giornaliero: {df['Volume'].mean():.0f}")  # calcolo la media della colonna Volume

# ESEGUIAMO
sp500 = load_data("drive/MyDrive/Progetto 2/sp500.csv")
euro50 = load_data("drive/MyDrive/Progetto 2/euro50.csv")

analyze_index(sp500, "S&P 500")
analyze_index(euro50, "EURO STOXX 50")
```
